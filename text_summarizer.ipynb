```python
from transformers import pipeline
import ipywidgets as widgets
from IPython.display import display, clear_output

```

```python
summarizer = pipeline("summarization")
```

```python
def summarize_text(text, max_len=100, min_len=30):
    summary = summarizer(text, max_length=max_len, min_length=min_len, do_sample=False)
    return summary[0]['summary_text']
```

```python
text_input = widgets.Textarea(
    value='Paste your long text here...',
    placeholder='Type or paste your text to summarize...',
    description='Input Text:',
    layout=widgets.Layout(width='100%', height='200px')
)
output = widgets.Output()

def on_text_change(change):
    with output:
        clear_output()
        if change['new'].strip():
            summary = summarize_text(change['new'])
            print("Summary:\n", summary)
        else:
            print("Please enter some text to summarize.")

text_input.observe(on_text_change, names='value')
```

```python
display(text_input, output)
```
